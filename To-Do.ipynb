```python
import tkinter as tk
from tkinter import messagebox

tasks = []

def add_task():
    task = task_entry.get()
    if task:
        tasks_listbox.insert(tk.END, task)
        tasks.append(task)  # Add task to the tasks list
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

def delete_task():
    selected_task_index = tasks_listbox.curselection()
    if selected_task_index:
        index = selected_task_index[0]
        tasks_listbox.delete(index)
        tasks.pop(index)  # Remove task from the tasks list

def update_task():
    selected_task_index = tasks_listbox.curselection()
    if selected_task_index:
        index = selected_task_index[0]
        new_task = task_entry.get()
        if new_task:
            tasks_listbox.delete(index)
            tasks_listbox.insert(index, new_task)
            tasks[index] = new_task

def delete_all_tasks():
    confirmed = messagebox.askyesno('Delete All', 'Do you want to delete all tasks?')
    if confirmed:
        tasks_listbox.delete(0, tk.END)
        tasks.clear()

# Create the main window
root = tk.Tk()
root.title("To-Do List")

# Create and place frames
functions_frame = tk.Frame(root)
functions_frame.pack(side=tk.LEFT, padx=20, pady=20)

listbox_frame = tk.Frame(root)
listbox_frame.pack(side=tk.RIGHT, padx=20, pady=20)

# Create and place widgets in functions frame
task_entry = tk.Entry(functions_frame, width=30)
task_entry.pack(pady=10)

add_button = tk.Button(functions_frame, text="Add Task", command=add_task)
add_button.pack()

delete_button = tk.Button(functions_frame, text="Delete Task", command=delete_task)
delete_button.pack()

update_button = tk.Button(functions_frame, text="Update Task", command=update_task)
update_button.pack()

delete_all_tasks_button = tk.Button(functions_frame, text='Delete All Tasks', command=delete_all_tasks)
delete_all_tasks_button.pack()

# Create and place list box in listbox frame
tasks_listbox = tk.Listbox(listbox_frame, selectmode=tk.SINGLE, width=40, height=10)
tasks_listbox.pack(pady=10)

# Start the main loop
root.mainloop()
```
